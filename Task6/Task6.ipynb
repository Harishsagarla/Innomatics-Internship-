#Problem 1: Smart Parking Lot Management **System**

```python
#Problem 1: Smart Parking Lot Management System
```

```python
def parking_system(capacity, logs):
    parked = 0
    peak = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        if parked > peak:
            peak = parked

    print("Currently Parked Vehicles:", parked)

    if parked > capacity:
        print("Parking Status: Full (Over Capacity Alert)")
    else:
        print("Parking Status: Available")

    print("Peak Usage:", peak)


# Sample Input
parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

#Problem 2: Online Food Delivery Time **Estimator**

```python
def delivery_eta(distance, traffic, weather):
    base_time = distance * 5   # 5 minutes per km

    if traffic == "High":
        base_time += 15
    elif traffic == "Medium":
        base_time += 10

    if weather == "Rainy":
        base_time += 10
    elif weather == "Storm":
        base_time += 20

    print("Estimated Delivery Time:", base_time, "minutes")


# Sample Input
delivery_eta(8, "High", "Rainy")
```

#Problem 3: Movie Theatre Seat Occupancy Analyzer

```python
def theatre_occupancy(total_seats, booked_list):
    booked = len(booked_list)
    occupancy = (booked / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Available")


# Sample Input
theatre_occupancy(200, [1] * 150)
```

#Problem 4: Cloud Server Load Classification System

```python
def server_load(cpu_readings):
    total = 0

    for cpu in cpu_readings:
        total += cpu

    avg = total / len(cpu_readings)

    print("Average CPU Load:", int(avg), "%")

    if avg < 50:
        print("Server Status: Normal")
    elif avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
server_load([45, 60, 70, 85, 90])
```

#Problem 5: Smart Classroom Resource Usage Monitor

```python
def classroom_monitor(resources):
    overused = []
    alert = False

    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)
            alert = True

    print("Overused Resources:", ", ".join(overused) if overused else "None")

    if alert:
        print("Energy Alert: Yes")
    else:
        print("Energy Alert: No")


# Sample Input
classroom_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

#Problem 6: Online Event Registration Capacity Controller

```python
def event_registration(capacity, registrations):
    if registrations <= capacity:
        confirmed = registrations
        waitlist = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)


# Sample Input
event_registration(100, 105)
```
